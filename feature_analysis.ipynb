```python
f, axarr = plt.subplots(20,3, figsize=(20, 80))
axarr = axarr.reshape(-1)
for i, n in enumerate(df_train.columns):
    df_train[n].value_counts().plot(kind='bar', ax=axarr[i], title=n)
```

```python
binary_features = [x for x in df_train.columns if x.endswith('_bin')]
```

```python
len(binary_features)
```

```python
f, axarr = plt.subplots(6,3, figsize=(20, 20))
axarr = axarr.reshape(-1)
for i, n in enumerate(binary_features):    
    data = df_train.groupby(by='target')[n].apply(lambda x: x.value_counts(True)).reset_index().rename(columns={'level_1': 'feature_val'})
    sns.barplot(x='feature_val', hue='target', y=n, data=data, ax=axarr[i])
    # plt.axes(axarr[i])
    # plt.title(n)

```
